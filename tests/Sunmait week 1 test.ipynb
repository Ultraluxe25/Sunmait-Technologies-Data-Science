### В чем разница между параметрами и гиперпараметрами?
Параметры настраиваются моделью во время обучения, этим занимается оптимизатор, высчитывая вектор-градиент по всем признакам. Гиперпараметры подбираются вручную Data Scientist-ом во время работы с отложенной (валидационной) выборкой или через один из видов кросс-валидации (например k-fold). Чаще всего подбор гиперпараметров осуществляется использую библиотеки optuna или hyperopt, я предпочитаю optuna.

### Какой критерий оптимизации используется в логистической регрессии?
В качестве критерия оптимизации в логистической регрессии используется метод максимального правдоподобия, который подбирает наиболее вероятные значения параметров на основании наблюдаемых данных. Для максимизации правдоподобия в логистической регрессии для задач классификации используется функция потерь бинарная кросс-энтропия, которая вычисляет расхождение значений классов с их предсказанными вероятностями. Логарифм в формуле BCE используется для сведения произведения в сумму (её легче считать и быстрее), а впереди стоящий минус отвечает за минимизацию функции потерь и в случае логарифма меньшего единицы делает результат применения функции положительным значением.

### Что такое глобальный минимум функции?
Глобальный минимум функции - минимальное значение функции на всей её области определения. Именно его стремятся достигнуть оптимизаторы, минимизируя функции потерь, но часто сходятся лишь к локальному минимуму (минимум на определенной интервале области определения функции).

### Чем отличается докер образ от докер контейнера?
Докер образ - это чертеж контейнера. Этот архив содержит в себе ОС, вашу программу, дополнительные библиотеки, настройки и команды запуска. Докер контейнер - это экземпляр конкретного докер образа (аналогия с классами и экземплярами в ООП). Если Docker Demon на хосте не находит образ, то он его скачивает с докер репозитория вроде DockerHub и на базе его создает контейнер.

### Напишите команду, которая запускает докер контейнер my_app, ассоциирует локальную папку /user с папкой /home в контейнере и ассоциирует локальный порт 8080 с портом 8000 контейнера
```docker run -v /user:/home -p 8080:8000 my_app```

**docker run** — запускает новый контейнер.

**-v /user:/home** — монтирует локальную папку /user в папку /home контейнера.

**-p 8080:8000** — перенаправляет локальный порт 8080 на порт 8000 контейнера.

**my_app** — имя Docker образа, на основе которого будет запущен контейнер.
